```python
import random
from random import sample
import time
import matplotlib.pyplot as plt
# seed random number generator
random.seed(1)
sequence = [i for i in range(100000)]
print(sequence)
```

```python
# select a subset without replacement
subset = sample(sequence, 50)
start = time.time()
for i in range(10000):
    subset.insert(50//5,25)
end = time.time()
runtime50 = end - start
print(f"Runtime of this snippet is {runtime50}")
#print(subset)
```
