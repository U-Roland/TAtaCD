```python
import random
from random import sample
import time
import matplotlib.pyplot as plt
```

```python
def measure(seq,number):
    start = time.time()
    subset = sample(seq, number)
    for i in range(10000):
        subset.insert(number//5,25)
    end = time.time()
    runtimeIn = end - start
    print(f"The runtime is {runtimeIn} sec")
    return runtimeIn
```

```python
random.seed(1)
sequence = [i for i in range(100000)]
seqRange = [50,100,500,1000,2000,5000,10000]
runtime = []
for i in range(7):
    print(f"{i+1}: for {seqRange[i]} operations\n --------------------")
    runtime.append(measure(sequence,seqRange[i]))
```

```python
plt.plot(seqRange,runtime)
```
